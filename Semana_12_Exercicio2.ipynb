```python
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar o dataset do Titanic
titanic = sns.load_dataset("titanic")

# 1. Filtrar apenas as colunas numéricas relevantes indicadas na imagem
colunas_relevantes = ["survived", "pclass", "age", "sibsp", "parch", "fare"]
df_filtrado = titanic[colunas_relevantes]

# Atenção: tratar valores ausentes com .dropna() antes de calcular a correlação
df_limpo = df_filtrado.dropna()

# Calcular a matriz de correlação
matriz_corr = df_limpo.corr()

# Mostrar os valores de correlação especificamente com a variável 'survived'
print("Correlação com a variável 'survived':")
print(matriz_corr["survived"].sort_values(ascending=False))

# 4. Experimentar as paletas "rocket" e "coolwarm" para comparação
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Heatmap com a paleta "rocket"
sns.heatmap(
    matriz_corr,
    annot=True,
    cmap="rocket",
    fmt=".2f",
    ax=axes[0],
    vmin=-1,
    vmax=1,
)
axes[0].set_title('Heatmap - Paleta "rocket"')

# Heatmap com a paleta "coolwarm"
sns.heatmap(
    matriz_corr,
    annot=True,
    cmap="coolwarm",
    fmt=".2f",
    ax=axes[1],
    vmin=-1,
    vmax=1,
)
axes[1].set_title('Heatmap - Paleta "coolwarm"')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar dataset
df = sns.load_dataset("tips")

# Calcular matriz
corr = df.corr(numeric_only=True)

# Criar uma máscara para o triângulo superior
mask = np.triu(np.ones_like(corr, dtype=bool))

# Plotar o heatmap com a máscara
sns.heatmap(
    corr,
    mask=mask,
    annot=True,
    cmap="coolwarm",
    center=0,
    linewidths=0.5,
    cbar_kws={"shrink": 0.8},
)

plt.title("Matriz de Correlação Filtrada – Dataset Tips")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Carregar o dataset iris
df = sns.load_dataset("iris")

# 2. Configurar o tamanho da figura (opcional, mas bom para relatórios)
plt.figure(figsize=(8, 5))

# 3. Criar o gráfico de dispersão (Scatter Plot)
sns.scatterplot(
    data=df,
    x="petal_length",  # Eixo X: Comprimento da pétala
    y="petal_width",  # Eixo Y: Largura da pétala
    hue="species",  # Separa e colore os pontos por espécie
    palette={
        "setosa": "purple",
        "versicolor": "tab:green",
        "virginica": "tab:blue",
    },  # Cores aproximadas da imagem
)

# 4. Customizar os rótulos e títulos em português
plt.title("Medidas das flores de íris")
plt.xlabel("Comprimento da pétala (cm)")
plt.ylabel("Largura da pétala (cm)")

# 5. Exibir o gráfico
plt.show()
```
