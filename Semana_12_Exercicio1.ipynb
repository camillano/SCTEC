```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Exercício 1
# Carregar o dataset
dataset_voos = sns.load_dataset("flights")

# Visualizar as primeiras linhas
dataset_voos.head()
```

```python
plt.figure(figsize=(12,5))

sns.lineplot(
    data=dataset_voos,
    x="year",
    y="passengers",
    errorbar=None
)

plt.title("Número de passageiros ao longo dos anos")
plt.xlabel("Ano")
plt.ylabel("Passageiros")

plt.show()
```

Existe uma tendência de crescimento dos passageiros ao longo dos anos.
Há oscilações mensais, mas o crescimento geral é evidente.

# Exercicio 2 - Grafico de barras

```python
import seaborn as sns
import matplotlib.pyplot as plt

dataset_pinguim = sns.load_dataset("penguins")

dataset_pinguim.head()
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=dataset_pinguim,
    x="species",
    y="body_mass_g"
)

plt.title("Massa corporal média por espécie")
plt.xlabel("Espécie")
plt.ylabel("Massa corporal (g)")

plt.show()
```

Resposta

A espécie Gentoo apresenta a maior massa corporal média.

Exercício 3 – Gráfico de dispersão

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=dataset_pinguim,
    x="flipper_length_mm",
    y="body_mass_g",
    hue="species"
)

plt.title("Comprimento da nadadeira x Massa corporal")
plt.xlabel("Comprimento da nadadeira (mm)")
plt.ylabel("Massa corporal (g)")

plt.show()
```

Existe uma correlação positiva:

quanto maior o comprimento da nadadeira;
maior tende a ser a massa corporal do pinguim.

Além disso, as espécies aparecem formando grupos distintos.
